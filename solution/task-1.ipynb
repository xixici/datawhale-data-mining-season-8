# task1 数据预处理分析

```python
import pandas as pd
data = pd.read_csv("../data.csv",encoding='gbk')
data.head()
```

# 数据信息

```python
data.info()
```

# 检查数据，drop无关特征

```python
drop_columns = ["Unnamed: 0", "custid","trade_no","bank_card_no","first_transaction_time","source","id_name","latest_query_time","loans_latest_time"]
data_drop_1 = data.drop(drop_columns, axis=1)
data_drop_1.info()
data_drop_1.head()
```

# 类型转换
- 转换object为int

```python
data_type_2 = pd.get_dummies(data_drop_1,columns=['reg_preference_for_trad'])
data_type_2.head()
```

# 缺失值处理
- 观察数据特征，删除含有缺失值的特征
- 插值方法填充缺失值
    - 均值
    - 众数
    - 特殊值

```python
# student有特殊意义，空值填充0
data_type_2["student_feature"].fillna(0, inplace=True)
# 其余特征众数填充即可
data_drop_cols = data_type_2.columns.values
for data_col in data_drop_cols:
    fill_value = data_type_2[data_col].value_counts().index[0]
    data_type_2[data_col].fillna(fill_value, inplace=True)
data_type_2.info()
data_type_2.head()
```
